```python
from googletrans import Translator
from tkinter import *
from tkinter import messagebox

# Initialize the Translator
translator_object = Translator()

# Define the translation function
def translate_function():
    # Retrieve source and destination language codes
    src_v = src_entry.get("1.0", "end-1c").strip().lower() or None
    dest_v = dest_entry.get("1.0", "end-1c").strip().lower() or None
    text_v = text_entry.get("1.0", "end-1c").strip()

    # Check if the text is empty
    if not text_v:
        messagebox.showerror("Error", "Enter valid text")
        return

    try:
        # Perform translation based on provided source and destination languages
        if src_v and dest_v:
            translated_text = translator_object.translate(text_v, src=src_v, dest=dest_v)
        elif src_v:
            translated_text = translator_object.translate(text_v, src=src_v)
        elif dest_v:
            translated_text = translator_object.translate(text_v, dest=dest_v)
        else:
            translated_text = translator_object.translate(text_v)

        # Display the translated text
        messagebox.showinfo("Translation", f"TRANSLATED TEXT: {translated_text.text}")

    except Exception as e:
        messagebox.showerror("Translation Error", f"Error: {e}")

# Define the clear function to reset all text fields
def clear():
    src_entry.delete("1.0", "end-1c")
    dest_entry.delete("1.0", "end-1c")
    text_entry.delete("1.0", "end-1c")

# Initialize main window
window = Tk()
window.geometry("500x300")
window.title("PythonGeeks Language Translator Using Python")

# Title Label
title_label = Label(window, text="PythonGeeks Language Translator Using Python", font=("Gayathri", 12))
title_label.pack()

# Text Input Label and Entry
text_label = Label(window, text="Text to translate:")
text_label.place(x=10, y=20)
text_entry = Text(window, width=40, height=5, font=("Ubuntu Mono", 12))
text_entry.place(x=130, y=20)

# Source Language Label and Entry
src_label = Label(window, text="Source language (empty: auto-detect):")
src_label.place(x=10, y=120)
src_entry = Text(window, width=20, height=1, font=("Ubuntu Mono", 12))
src_entry.place(x=275, y=120)

# Destination Language Label and Entry
dest_label = Label(window, text="Target language (empty: English-default):")
dest_label.place(x=10, y=150)
dest_entry = Text(window, width=20, height=1, font=("Ubuntu Mono", 12))
dest_entry.place(x=275, y=150)

# Translate and Clear Buttons
button_translate = Button(window, text='Translate', bg='Turquoise', command=translate_function)
button_translate.place(x=160, y=190)

button_clear = Button(window, text='Clear', bg='Turquoise', command=clear)
button_clear.place(x=270, y=190)

# Run the main event loop
window.mainloop()
```
